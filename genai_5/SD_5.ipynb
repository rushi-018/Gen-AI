```python
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
```

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
).to("cuda")
```

```python
prompt = "A surreal landscape with floating islands, glowing waterfalls, and futuristic temples under a starry sky"
```

```python
image = pipe(prompt).images[0]
```

```python
image.show()
image.save("creative_artwork.png")
```
