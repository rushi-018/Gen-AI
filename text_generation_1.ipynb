```python
import random

text = (
    "Artificial intelligence is transforming the world "
    "Artificial intelligence is used in healthcare education and finance "
    "Artificial intelligence helps machines learn from data "
    "Machine learning is a part of artificial intelligence "
    "Deep learning is a type of machine learning "
    "Artificial intelligence improves efficiency and accuracy "
)

words = text.split()

markov = {}
for i in range(len(words) - 1):
    if words[i] not in markov:
        markov[words[i]] = []
    markov[words[i]].append(words[i + 1])

current = random.choice(list(markov.keys()))
result = [current]

for i in range(30):
    current = random.choice(markov[current])
    result.append(current)

print(" ".join(result))
```
