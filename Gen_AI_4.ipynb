Now that we have the `transformers` library installed, I will load a pre-trained question-answering model and create a pipeline. We'll use `distilbert-base-cased-distilled-squad` which is a good balance of performance and speed.

```python
from transformers import pipeline

# Load a pre-trained question-answering model
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

print("Question-answering pipeline loaded successfully!")
```

Next, let's define a context from which our chatbot will answer questions. You can replace this with any text you want the chatbot to read.

```python
context = """\nGoogle Colaboratory, commonly referred to as Colab, is a free cloud-based Jupyter notebook environment provided by Google. It allows users to write and execute Python code through a web browser, and is especially well-suited for machine learning, data analysis, and education. Colab provides free access to computing resources including GPUs and TPUs, making it a popular choice for training deep learning models without the need for expensive local hardware. It integrates seamlessly with Google Drive, allowing for easy saving and sharing of notebooks. Users can also install various libraries and frameworks, customize their environment, and collaborate with others on projects.\n"""

print("Context defined successfully.")
```

Now, let's implement the interactive chatbot. You can ask questions based on the provided context.

```python
def chatbot_qa():
    print("\nWelcome to the QA Chatbot! Ask me anything about the provided context.\n")
    print("Type 'quit' to exit.")

    while True:
        question = input("\nYour Question: ")
        if question.lower() == 'quit':
            print("Exiting chatbot. Goodbye!")
            break

        if not question:
            print("Please enter a question.")
            continue

        try:
            result = qa_pipeline(question=question, context=context)
            print(f"Answer: {result['answer']}")
            print(f"Score: {result['score']:.2f}")
        except Exception as e:
            print(f"An error occurred: {e}")

# Run the chatbot
chatbot_qa()
```
